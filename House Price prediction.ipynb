# Scaling and Train Test Split

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

**Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Create a model

```python
import tensorflow as tf
```

```python
epochs=10
batch_size = 32
```

```python
# normalize all the data points and cast the labels to int64
X_train, X_test = tf.cast(X_train, tf.float32), tf.cast(X_test, tf.float32)
y_train, y_test = tf.cast(y_train,tf.int64),tf.cast(y_test,tf.int64)
```

```python
#house_model = tf.keras.models.Sequential([
#tf.keras.layers.Dense(20,activation=tf.nn.relu),
#tf.keras.layers.Dropout(0.2),
#tf.keras.layers.Dense(20,activation=tf.nn.relu),
#tf.keras.layers.Dense(20,activation=tf.nn.relu),
#tf.keras.layers.Dropout(0.2),
#tf.keras.layers.Dense(20,activation=tf.nn.relu),
#tf.keras.layers.Dense(1,activation=tf.nn.softmax)
#])
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

house_model = Sequential()
house_model.add(Dense(19,activation='relu'))
house_model.add(Dense(19,activation='relu'))
house_model.add(Dense(19,activation='relu'))
house_model.add(Dense(19,activation='relu'))
house_model.add(Dense(1))
```

```python
optimiser = tf.keras.optimizers.Adam()
house_model.compile (optimizer= optimiser, loss='mse')
```

```python
house_model.fit(X_train, y_train, batch_size=128, validation_data = (X_test, y_test), epochs=400)
#model.fit(x=X_train,y=y_train.values,
#          validation_data=(X_test,y_test.values),
#          batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(house_model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = house_model.predict(X_test)
```
