**Q)Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()**

```python
def myreduce(af, seq):
    result = seq[0]
    for i in seq[1:]:
        result = af(result, i)
    return result
```

```python
# Test myreduce() function
def sum(x,y): return x + y
```

```python
print ("Sum of the list [45, 50, 55] using 'myreduce' function is: "   + str(myreduce(sum, [45,50,55])))
```

**Q)Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()**

```python
def myfilter(af, seq):
    result = []
    for i in seq:
        if af(i):
            result.append(i)
    return result
```

```python
# Test myfilter() function
def ispositive(x):
    if (x <= 0): 
        return False 
    else: 
        return True
```

```python
ispositive(0)
```

```python
print ("Filter only positive Integers on list [0,453,-853,958,882,225] using 'myfilter()' function"  + str(myfilter(ispositive, [0,453,-853,958,882,225])))
```

**Q1.1 - Write a Python Program(with class concepts) to find the area of the triangle using the below formula.**

**area = (s*(s-a)*(s-b)*(s-c)) ** 0.5**

**Function to take the length of the sides of triangle from user should be defined in the parentclass and function to calculate 
the area should be defined in subclass.**

```python
import math

class triangle:
    x = 0
    y = 0
    z = 0
    s = 0
    
    def __init__(self, x,y,z):
        self.x = x
        self.y = y
        self.z = z
        self.s = (x+y+z) / 2
        
    def area(self):
        area = math.sqrt(self.s*(self.s - self.x) * (self.s - self.y) * (self.s - self.z))
        return area
    
tri = triangle(6,7,8)
print(tri.area())
```

**Q1.2 - Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that 
are longer than n.**

```python
def filter_long_words(words, n):
    return list(filter(lambda x: (len(x) ) > n, words))
print(filter_long_words(["afiudhcvodhfoac", "udfhuefeyeyyyeyebvavu", "yrte", "uu"], 8))
```

**Q1.3 - Write a Python program using function concept that maps list of words into a list of integers representing the lengths 
of the corresponding words.**

**Hint: If a list [ ab,cde,erty] is passed on to the python function output should come as [2,3,4]. Here 2,3 and 4 are the 
lengths of the words in the list.**

```python
def word_len(words):
    return list(map(lambda x: len(x), words))

print(word_len(["dster",'uhfufe']))
```

**Q1.4 - Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False 
otherwise.**

```python
def is_vovel(char):
    if len(char) == 1 and char in ["a", "e", "i", "o", "u"]:
        return True
    else:
        return False

print(is_vovel('r'))
```
